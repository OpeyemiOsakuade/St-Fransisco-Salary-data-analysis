```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Importing the data
data = pd.read_csv ('Salaries.csv')
data.head()
```

```python
#to make python see 'not provided' in the Employeename as a missing value
missing_values = ["Not provided"]
data = pd.read_csv("Salaries.csv", na_values = missing_values)
```

```python
data
```

```python
#FILLED THE EMPTY SPACES IN benefits,basepay,overtimepay and other pay with zero so the python does not see it as a missing value.
data[['Benefits','BasePay','OvertimePay','OtherPay']] = data[['Benefits','BasePay','OvertimePay','OtherPay']].fillna(value=0)
data
```

```python
#Deleted 'Notes','Status','Agency' since they are not needed
data.drop(['Notes','Status','Agency','Id'], axis=1, inplace = True)
```

```python
data
```

```python
data.dropna() #deleted all missing values(NaN)
```

```python
#To determine how the workers are paid per year
plt.figure(figsize = (8,8))
plt.title("Worker's Basepay per year", FontSize =15)
sns.barplot(x='Year', y='BasePay', data=data);
```

```python
#To determine how the workers get additional Benefits per year
plt.figure(figsize = (8, 8))
plt.title("Worker's Benefits per year", FontSize =15)
sns.barplot(x='Year', y='Benefits', data=data);
```

```python
#Shows how the over time pay to workers increases per year
plt.figure(figsize = (8,8)) 
sns.barplot(x='Year', y='OvertimePay', data=data)
plt.title("Worker's OvertimePay per year", FontSize =15)
```

```python
#To determine how the workers get other pays per year #paid by rank inrespect to basepay
plt.figure(figsize = (8,8)) 
sns.barplot(x='Year', y='OtherPay', data=data)
plt.title("Worker's OtherPay per year", FontSize =1)
```

```python
# aggregate by job title and pick out the BasePay, Benefits, and Overtime features
salaries_averages_by_occupation = (data[data.JobTitle.isin(top_ten_occupations)]
                                   .groupby('JobTitle')[['BasePay', 'Benefits', 'OvertimePay','OtherPay']]
                                   .aggregate('mean')
)

ax = salaries_averages_by_occupation.plot(kind='barh', figsize=(8,8))

ax.set_xlabel('Mean Pay')
```
